```python
import pandas as pd
```

```python
file = "data/goal_9.xlsx"
```

```python
regions_to_remove = [
    "Africa", "Americas", "Asia", "Australia and New Zealand", "Caribbean",
    "Caucasus and Central Asia", "Central America", "Central Asia",
    "China, Hong Kong Special Administrative Region", "China, Macao Special Administrative Region",
    "Developed regions (Europe, Cyprus, Israel, Northern America, Japan, Australia & New Zealand)",
    "Developing regions", "Eastern Africa", "Eastern and South-Eastern Asia",
    "Eastern Asia", "Eastern Asia (excluding Japan and China)",
    "Eastern Asia (excluding Japan)", "Eastern Europe", "Europe",
    "Europe and Northern America", "Landlocked developing countries (LLDCs)",
    "Latin America and the Caribbean", "Least Developed Countries (LDCs)", "Melanesia",
    "Middle Africa", "Northern Africa", "Northern Africa (exc. Sudan)",
    "Northern Africa and Western Asia", "Northern America", "Northern Europe",
    "Oceania", "Oceania (exc. Australia and New Zealand)",
    "Other non-specified areas in Eastern Asia", "Small island developing States (SIDS)",
    "South America", "South-Eastern Asia", "Southern Africa", "Southern Asia",
    "Southern Asia (excluding India)", "Southern Europe", "Sub-Saharan Africa",
    "Sub-Saharan Africa (inc. Sudan)", "Western Africa", "Western Asia",
    "Western Asia (exc. Armenia, Azerbaijan, Cyprus, Israel and Georgia)", "Western Europe",
    "World"
]
```

```python
df = pd.read_excel(file, sheet_name="Goal9 Cleaned")
df.head()
```

```python
filtered_df = df[~df['GeoAreaName'].isin(regions_to_remove)]

filtered_df.head()
```

```python
with pd.ExcelWriter(file, mode='a', engine='openpyxl') as writer:
    filtered_df.to_excel(writer, index=False, sheet_name="Goal9 Filtered")
```
